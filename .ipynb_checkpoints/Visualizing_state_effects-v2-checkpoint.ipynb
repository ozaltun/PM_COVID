```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date_of_study = "05-19-2020"
save_root = "/Users/ozaltun/Dropbox (MIT)/Project_coal_ng_mort/Notes/Harvard paper/{}"
df = pd.read_csv("/Users/ozaltun/Dropbox (MIT)/Data/Health/covid/"+date_of_study+\
                 "_estimates.csv", index_col=0)
```

```python
df.date = df.date.str.slice(start=1)
df['date'] = pd.to_datetime(df['date'])
#df['date'] = df.date.dt.strftime('%Y-%m-%d')
```

```python
df["ci_indicator"] = np.sign(df.lb) + np.sign(df.ub)
df["ci_indicator"] = df["ci_indicator"]/2
df["ci_indicator"] = df["ci_indicator"].astype(int)
df.head()
```

```python
df.cf_name.unique()
```

```python
coefficients = df.cf_name.unique()
states = [i for i in coefficients if 'factor(state)' in i]
states_map = {i:i[13:]for i in states}

coef_subset = ['mean_pm25', 'share.drive','share.PublicTransit','share.Home','percent.smokers',\
               'percent.obese','percent.uninsured','percent.obese','older_pecent','pct_blk',\
              'log(medianhousevalue)']
coef_subset_map = {'mean_pm25':'PM2.5', 'share.drive':'Share - driving','share.PublicTransit':'Share - public transit',\
                   'share.Home':'Share - home','percent.smokers':'Percent smokers','percent.obese':'Percent Obese',\
                   'percent.uninsured':'Percent Uninsured','percent.obese':'Percent Obese',\
                   'older_pecent':'Percent old','pct_blk':'Percent Black', 'log(medianhousevalue)':'log(medianhousevalue)'}
other_coefs = [i for i in coefficients if (i not in states) & (i not in coef_subset)]
other_coefs_map = {i:i for i in other_coefs}
coef_map = {**states_map, **coef_subset_map, **other_coefs_map}
df.cf_name = df.cf_name.map(coef_map)
df.head()
```

```python
states = list(states_map.values())
coef_subset = list(coef_subset_map.values())
print(states)
print(coef_subset)
```

```python
dates = df.date.unique()
selected_dates = {'Early':dates[0], 'Mid':dates[int(dates.shape[0]/2)],'Latest':dates[dates.shape[0]-1]}
selected_dates
```

```python
df_levels = df.set_index(['type','date','cf_name'])
```

```python
# df_levels.loc['with_states',:,states][df_levels.ci_indicator !=0]
```

```python
fig, (ax, ax1) = plt.subplots(1,2, figsize=(30,15))

# All
df_temp = df_levels.loc['with_states',:,states].reset_index().drop(columns=['type'])
ordered_states = df_temp[df_temp.date == '2020-05-16'].sort_values(by=['est'], ascending=False).cf_name.values
for i, state in enumerate(ordered_states):
    df_temp_state = df_temp[df_temp.cf_name == state].sort_values(by=['date'])
    ax.plot(df_temp_state['date'], df_temp_state['est'], label=state)
ax.set_xticks(dates[::5])
ax.legend(loc='upper left')
ax.set_xlabel('Day')
ax.set_ylabel('Coefficient')

# Only statistically significant
df_temp = df_levels.loc['with_states',:,states].reset_index().drop(columns=['type'])
df_temp = df_temp[df_temp.ci_indicator !=0]
ordered_states = df_temp[df_temp.date == '2020-05-16'].sort_values(by=['est'], ascending=False).cf_name.values
for i, state in enumerate(ordered_states):
    df_temp_state = df_temp[df_temp.cf_name == state].sort_values(by=['date'])
    ax1.plot(df_temp_state['date'], df_temp_state['est'], label=state)
ax1.set_xticks(dates[::5])
ax1.legend(loc='upper left')
ax1.set_xlabel('Day')
ax1.set_ylabel('Coefficient')
fig.savefig(save_root.format('line_states.png'), dpi=fig.dpi)
plt.show()
```

```python
rows = 5
fig, ax = plt.subplots(rows,2, figsize=(30,30))

# All
df_temp = df_levels.loc['with_states',:,coef_subset].reset_index().drop(columns=['type'])
ordered_coefs = df_temp[df_temp.date == '2020-05-16'].sort_values(by=['est'], ascending=False).cf_name.values
for i, coef in enumerate(ordered_coefs):
    df_temp_coefs = df_temp[df_temp.cf_name == coef].sort_values(by=['date'])
    ax[i %rows, int(i/rows)].plot(df_temp_coefs['date'], df_temp_coefs['est'], label=coef)
    ax[i %rows, int(i/rows)].fill_between(df_temp_coefs['date'], df_temp_coefs['lb'], df_temp_coefs['ub'],alpha=0.2)
    ax[i %rows, int(i/rows)].axhline(y=0, color='r', linestyle='--')
    ax[i %rows, int(i/rows)].set_xticks(dates[::5])
    ax[i %rows, int(i/rows)].legend(loc='upper left')
    ax[i %rows, int(i/rows)].set_xlabel('Day')
    ax[i %rows, int(i/rows)].set_ylabel('Coefficient')
    
    
fig.savefig(save_root.format('line_coefs.png'), dpi=fig.dpi)
plt.show()
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(30,15))
# Only statistically significant
df_temp = df_levels.loc['with_states',:,coef_subset].reset_index().drop(columns=['type'])
df_temp = df_temp[df_temp.ci_indicator !=0]
dates_subset = df_temp.date.unique()
# ordered_coefs = df_temp[df_temp.date == '2020-05-16'].sort_values(by=['est'], ascending=False).cf_name.values
for i, coef in enumerate(ordered_coefs):
    df_temp_coefs = df_temp[df_temp.cf_name == coef].sort_values(by=['date'])
    ax1.plot(df_temp_coefs['date'], df_temp_coefs['est'],'-*', label=coef)
ax1.legend(loc='upper left')
ax1.set_xticks(dates_subset[::5])
#ax1.set_xlim([datetime.date(2020, 4, 4), datetime.date(2020, 5, 16)])
#ax1.set_xlim([dates[0], dates[-1]])
ax1.set_xlabel('Day')
ax1.set_ylabel('Coefficient')
fig.savefig(save_root.format('line_sig_coefs.png'), dpi=fig.dpi)
plt.show()
```

```python
date_for_bar_graph = dates[-1]
#df_temp = df[(df.type == 'with_states') & (df.date ==date_for_bar_graph)].drop(columns=['type', 'date'])
df_temp = df_levels.loc['with_states',date_for_bar_graph,states].reset_index().drop(columns=['type']).sort_values(by=['est'], ascending=False)

fig, ax = plt.subplots(1,1, figsize=(40,20))
ax.bar(df_temp['cf_name'], df_temp['est'], yerr=(df_temp['ub']-df_temp['lb'])/2, align='center', alpha=0.5, ecolor='black', capsize=10)
# ax.set_ylabel('Coefficient of Thermal Expansion ($\degree C^{-1}$)')
ax.set_xticklabels(df_temp['cf_name'], rotation=65)
# ax.set_title('Coefficent of Thermal Expansion (CTE) of Three Metals')
# ax.yaxis.grid(True)
fig.savefig(save_root.format('bar_states.png'), dpi=fig.dpi)
# Save the figure and show
plt.show()
```
